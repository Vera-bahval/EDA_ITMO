```python
import pandas as pd
```

```python
transactions = pd.read_parquet('transaction_fraud_data.parquet')
currency_rates = pd.read_parquet('historical_currency_exchange.parquet')
```

```python
# Вопрос 1: Какая доля всех транзакций — мошеннические?
fraud_percentage = transactions['is_fraud'].mean().round(1)
print(f"1. Доля мошеннических транзакций: {fraud_percentage}")
```

```python
# Вопрос 2: Топ-5 стран по мошенническим транзакциям
top_fraud_countries = transactions[transactions['is_fraud']]['country'].value_counts().head(5).index.tolist()
print(f"2. Топ-5 стран: {','.join(top_fraud_countries)}")
```

```python
# Вопрос 3: Сколько транзакций в среднем совершает один клиент за час?
transactions['hour'] = transactions['timestamp'].dt.floor('H')

client_hour_trans = transactions.groupby(['customer_id', 'hour']).size().reset_index(name='transaction_count')

avg = client_hour_trans['transaction_count'].mean()
print(f"3. Кол-во транзакций в час на одного клиента: {avg:.2f}")
```

```python
# Вопрос 4: Доля мошенничества у high-risk продавцов
high_risk_fraud = transactions[transactions['is_high_risk_vendor']]['is_fraud'].mean().round(1)
print(f"4. Доля мошенничества у high-risk продавцов: {high_risk_fraud}")
```

```python
# Вопрос 5: В каком городе наибольшая средняя сумма транзакций?
import numpy as np

# словарь средних курсов валют
avg_rates = currency_rates.select_dtypes(include=[np.number]).mean().to_dict()
avg_rates['USD'] = 1.0

def convert_to_usd(amount, currency):
    if currency == 'USD':
        return amount
    elif currency in avg_rates:
        return amount / avg_rates[currency]
    else:
        return amount

transactions['amount_usd'] = transactions.apply(
    lambda row: convert_to_usd(row['amount'], row['currency']), axis=1
)

city_avg = transactions.groupby('city')['amount_usd'].agg(['mean', 'count']).reset_index()
city_avg.columns = ['city', 'avg_amount_usd', 'transaction_count']

city_avg = city_avg.sort_values('avg_amount_usd', ascending=False)

top_city = city_avg.iloc[0]['city']
top_avg = city_avg.iloc[0]['avg_amount_usd']

print(f"5. Город с наибольшей транзакцией: {top_city}")
print(f"Средняя сумма: ${top_avg:.2f} USD")
```

```python
# Вопрос 6: В каком городе выше всего средний чек по операциям, связанным с fast_food?
fast_food_df = transactions[transactions['vendor_type'] == 'fast_food'].copy()

fast_food_df['amount_usd'] = fast_food_df.apply(
    lambda row: convert_to_usd(row['amount'], row['currency']), axis=1
)

city_avg_check = fast_food_df.groupby('city')['amount_usd'].agg(['mean', 'count']).reset_index()
city_avg_check.columns = ['city', 'avg_check_usd', 'transaction_count']

city_avg_check = city_avg_check.sort_values('avg_check_usd', ascending=False)

top_city = city_avg_check.iloc[0]['city']
top_avg = city_avg_check.iloc[0]['avg_check_usd']

print(f"Город с наибольшим чеком fast-food: {top_city}")
print(f"Средний чек: ${top_avg:.2f} USD")
```

```python
# Вопрос 7: Среднее немошеннических операций в USD
import math

non_fraud = transactions[transactions['is_fraud'] == False]
avg_rates = currency_rates.mean(numeric_only=True)

currency_groups = non_fraud.groupby('currency')['amount'].sum()

total_usd = 0
total_count = 0

for currency, total_amount in currency_groups.items():
    count = (non_fraud['currency'] == currency).sum()
    
    if currency == 'USD':
        usd_amount = total_amount
    elif currency in avg_rates.index:
        usd_amount = total_amount / avg_rates[currency]
    else:
        usd_amount = total_amount
    
    total_usd += usd_amount
    total_count += count
    print(f"{currency}: {total_amount:.2f} -> {usd_amount:.2f} USD ({count} операций)")

avg_usd = total_usd / total_count
result = math.ceil(avg_usd)

print(f"7. Среднее немошеннических операций в USD: {result}")
```

```python
# Вопрос 7: STD немошеннических операций в USD
non_fraud = transactions[transactions['is_fraud'] == False].copy()

non_fraud['amount_usd'] = non_fraud['amount'] / non_fraud['currency'].map(avg_rates).fillna(1.0)

std_usd = non_fraud['amount_usd'].std()
result = math.ceil(std_usd)

print(f"7. STD немошеннических операций в USD: {result}")
```

```python
# Вопрос 8: Среднее мошеннических операций в USD
import numpy as np

fraud_df = transactions[transactions['is_fraud'] == True]

avg_rates = currency_rates.select_dtypes(include=[np.number]).mean().to_dict()
avg_rates['USD'] = 1.0

fraud_df = fraud_df.copy()
fraud_df['amount_usd'] = fraud_df.apply(
    lambda row: row['amount'] if row['currency'] == 'USD' 
    else row['amount'] / avg_rates.get(row['currency'], 1.0), axis=1
)

mean_fraud_usd = fraud_df['amount_usd'].mean()

print(f"8. Среднее мошеннических операций в USD {mean_fraud_usd:.2f}")
```

```python
# Вопрос 9: STD мошеннических операций в USD
fraud_df['amount_usd'] = fraud_df['amount'] / fraud_df['currency'].map(avg_rates).fillna(1.0)

std_usd = fraud_df['amount_usd'].std()
result = math.ceil(std_usd)

print(f"9. STD мошеннических операций в USD: {result}")
```

```python
# Вопрос 10: Клиенты с опасным поведением
transactions['unique_merchants'] = transactions['last_hour_activity'].apply(lambda x: x['unique_merchants'])
median_merchants = transactions.groupby('customer_id')['unique_merchants'].median()
threshold = median_merchants.quantile(0.95)
dangerous_clients = (median_merchants > threshold).sum()
print(f"10. Клиентов с опасным поведением: {dangerous_clients}")
```
